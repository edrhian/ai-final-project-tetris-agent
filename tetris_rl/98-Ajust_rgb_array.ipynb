```python
from tetris_env import TetrisEnv
import random
import numpy as np
import matplotlib.pyplot as plt
from IPython import display
from gymnasium.wrappers import FrameStackObservation
from gymnasium.wrappers import ResizeObservation
from gymnasium.wrappers import GrayscaleObservation
from gymnasium.wrappers import ReshapeObservation
from gymnasium.wrappers import TransformObservation

from tetris_env import TetrisEnv

from gymnasium.wrappers import TransformObservation
from PIL import Image
```

```python
def display_rgb(rgb_array):
    plt.imshow(rgb_array)
    plt.axis('off')  # Hide axes
    plt.show()
```

```python
import time
# left = 0
# right = 1
# rotate_clock = 2
# softdrop = 3
# hardrop = 4
# hold_piece = 5
# rotate_counter_clock = 6
# rotate_180 = 7

env = TetrisEnv(board_width=10,
                board_height=22,
                time_limit_s=120,
                obs_type="rgb_array",
                render_mode='human')

env = TransformObservation(env, lambda obs: obs[:, 0:300], observation_space=None)
env = GrayscaleObservation(env)
env = ResizeObservation(env, (84, 84))

print(env.render_mode)
n_actions = env.action_space.n
state, info = env.reset(seed=27)
print(state.shape)
done = False
for i in range(5):
    action = random.randint(0,7)
    action = 4
    
    observation, reward, terminated, truncated, info = env.step(action)
    done = truncated or terminated
    print(observation.shape)
    display_rgb(observation)
    # print(obs_one_hot(observation,env))
    time.sleep(0.07)
    display.clear_output(wait=True)
    img = Image.fromarray(observation, 'L')
    img.save("canvasx.png")
    # crop_obs = observation[:,:300]
    # Without ResizeObservation
    # crop_obs = observation[:,:300]
    # print(crop_obs.shape)
    # display_rgb(crop_obs)
    if done:
        break

env.close()
```
